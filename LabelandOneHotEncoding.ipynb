```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from google.colab import files
import os

# Step 1: Upload the dataset
uploaded = files.upload()

# Step 2: Verify the file exists
print("Files in directory:", os.listdir())

# Step 3: Load the dataset
df = pd.read_csv('Housing copy.csv')  # Use the exact filename from os.listdir()

# Step 4: Display unique values in 'furnishingstatus'
print("\nUnique furnishingstatus values:", df['furnishingstatus'].unique())

# Step 5: Label Encoding
label_encoder = LabelEncoder()
df['furnishingstatus_label'] = label_encoder.fit_transform(df['furnishingstatus'])
print("\nLabel Encoded 'furnishingstatus':\n", df[['furnishingstatus', 'furnishingstatus_label']].head(10))

# Step 6: One-Hot Encoding
df_onehot = pd.get_dummies(df, columns=['furnishingstatus'], drop_first=False)
print("\nOne-Hot Encoded columns:\n", df_onehot.filter(like='furnishingstatus').head(10))
```
