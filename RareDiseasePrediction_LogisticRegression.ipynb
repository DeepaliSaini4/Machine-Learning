```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, classification_report
from google.colab import files
import os

# Step 1: Upload the dataset
uploaded = files.upload()

# Step 2: Verify the file exists
print("Files in directory:", os.listdir())

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, classification_report

# Load the dataset using the correct filename
df = pd.read_csv('Disease_symptom_and_patient_profile_dataset.csv')

# Step 1: Define rare diseases (threshold = 10 cases)
disease_counts = df['Disease'].value_counts()
rare_diseases = disease_counts[disease_counts < 10].index
df['is_rare'] = df['Disease'].isin(rare_diseases).astype(int)

# Step 2: Convert 'Yes'/'No' in symptom columns to 1/0
symptom_cols = ['Fever', 'Cough', 'Fatigue', 'Difficulty Breathing']
for col in symptom_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0})

# Step 3: Encode categorical variables
df = pd.get_dummies(df, columns=['Gender', 'Blood Pressure', 'Cholesterol Level'], drop_first=True)

# Step 4: Prepare features and target
X = df.drop(['Disease', 'Outcome Variable', 'is_rare'], axis=1)
y = df['is_rare']

# Step 5: Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 6: Train logistic regression with class weighting
model = LogisticRegression(class_weight='balanced', max_iter=1000)
model.fit(X_train, y_train)

# Step 7: Predict and evaluate
y_pred = model.predict(X_test)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))

```
