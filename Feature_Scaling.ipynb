```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Upload the dataset
from google.colab import files
uploaded = files.upload()  # Upload Housing.csv when prompted

# Verify the file has been uploaded
import os
print("Files in directory:", os.listdir())

# Now proceed with your feature scaling code
housing_data = pd.read_csv('Housing copy.csv')

# Select numerical features only
numerical_features = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking']
data_for_scaling = housing_data[numerical_features]

# Display original data histograms
plt.figure(figsize=(15, 5))
for i, col in enumerate(data_for_scaling.columns):
    plt.subplot(1, 5, i+1)
    plt.hist(data_for_scaling[col], bins=10, color='skyblue', edgecolor='black')
    plt.title(f'Original {col}')
plt.tight_layout()
plt.show()

# Standardization (zero mean, unit variance)
scaler_std = StandardScaler()
housing_std = scaler_std.fit_transform(data_for_scaling)
housing_std_df = pd.DataFrame(housing_std, columns=data_for_scaling.columns)

# Normalization (min-max scaling to [0,1])
scaler_norm = MinMaxScaler()
housing_norm = scaler_norm.fit_transform(data_for_scaling)
housing_norm_df = pd.DataFrame(housing_norm, columns=data_for_scaling.columns)

# Display histograms after standardization
plt.figure(figsize=(15, 5))
for i, col in enumerate(housing_std_df.columns):
    plt.subplot(1, 5, i+1)
    plt.hist(housing_std_df[col], bins=10, color='lightgreen', edgecolor='black')
    plt.title(f'Standardized {col}')
plt.tight_layout()
plt.show()

# Display histograms after normalization
plt.figure(figsize=(15, 5))
for i, col in enumerate(housing_norm_df.columns):
    plt.subplot(1, 5, i+1)
    plt.hist(housing_norm_df[col], bins=10, color='salmon', edgecolor='black')
    plt.title(f'Normalized {col}')
plt.tight_layout()
plt.show()
```
